```python
employee_names = ['Muhammed','Essa','Hameed' ,'Hassan','Waleed' ]
employee_salary = [3000,7000,2000 ,8000,5000 ]

plt.figure(figsize=(35,4))

plt.subplot(133) 
plt.bar(employee_names,employee_salary  )

plt.subplot(132) 
plt.plot(employee_names,employee_salary  )

plt.subplot(131) 
plt.scatter(employee_names,employee_salary  )

plt.title("employee info")
plt.ylabel("salary axis")
plt.xlabel("names axis")
plt.show()
```
